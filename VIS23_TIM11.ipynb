**Pitanje 5:**
Koja bolest više utiče na skraćivanje životnog vijeka, HIV ili Hepatitis?

```python
import pandas as pd
import matplotlib.pyplot as plt
# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# Odabir samo potrebnih podataka za analizu
df = df[['Country', 'Status', 'Hepatitis B', ' HIV/AIDS', 'Life expectancy ']]

# Uklanjanje nedostajućih vrijednosti
df = df.dropna()

# Odabir zemalja za analizu
high_hiv = ['South Africa', 'Swaziland']
high_hep_b = ['Moldova', 'Romania']

# Filtriranje podataka samo za zemlje s visokom stopom zaraze HIV-om i zemlje s visokom stopom zaraze hepatitisom B
high_hiv_data = df.loc[(df[' HIV/AIDS'] >= 1) & (df['Country'].isin(high_hiv))]
high_hep_b_data = df.loc[(df['Hepatitis B'] >= 90) & (df['Country'].isin(high_hep_b))]

# Vizualizacija podataka
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
high_hiv_data.boxplot(column='Life expectancy ', by='Country', ax=ax1)
ax1.set_title('Visoka stopa zaraze HIV-om')
ax1.set_ylabel('Očekivano trajanje života')
ax2.set_ylim(ax1.get_ylim())  # Postavljanje istog raspona na osi y u oba boxplot-a
high_hep_b_data.boxplot(column='Life expectancy ', by='Country', ax=ax2)
ax2.set_title('Visoka stopa zaraze hepatitisom B')
ax2.set_ylabel('Očekivano trajanje života')
plt.suptitle('Uticaj HIV-a i hepatitisa B na očekivano trajanje života')
plt.show()

```

**Zaključak:** Na osnovu opšteg znanja o HIV-u i hepatitisu B, oba ova virusa mogu smanjiti životni vijek u proseku u zemljama u kojima su približno isto zastupljeni. Međutim, uticaj svakog virusa na smanjenje životnog vijeka može varirati u zavisnosti od mnogobrojnih faktora kao što su dostupnost i efikasnost tretmana, socijalno-ekonomski uslovi,  životni stil stanovništva itd. Stoga je teško precizno odrediti koji virus ima veći uticaj na smanjenje životnog vijeka čak iako su podaci približni. Sam primjer je vidljiv sa grafika gdje Južnoafrička Republika ima veće očekivano trajanje života od Kraljevine Svazi, iako su podaci u vezi HIV- prilično pribli

**Pitanje 6:**
Kako BMI utiče na dužinu životnog vijeka?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# Odabir samo potrebnih podataka za analizu
df = df[['Country', 'Year', 'Life expectancy ', ' BMI ']]

# Filtriranje podataka za odabrane zemlje
country_names = ['United States of America', 'Germany', 'India']
dfs = {}
for country_name in country_names:
    dfs[country_name] = df.loc[df['Country'] == country_name]

# Stvaranje scatter plot-ova za prikaz veze između BMI-a i očekivanog trajanja života za svaku zemlju
fig, axs = plt.subplots(nrows=len(country_names), ncols=1, figsize=(6, 8))
for i, country_name in enumerate(country_names):
    axs[i].scatter(dfs[country_name][' BMI '], dfs[country_name]['Life expectancy '])
    axs[i].set_xlabel(' BMI ')
    axs[i].set_ylabel('Očekivano trajanje života')
    axs[i].set_title('Utjecaj BMI-a na životni vijek u zemlji: ' + country_name)
plt.tight_layout()
plt.show()

```

**Zaključak:** Sa grafa se jasno vidi da u svakoj od zemalja uzorka, prosječno očekivano trajanje života se linearno povećava sa povećanjem BMI.

**Pitanje 11:** Postoji li veza između stope smrtnosti djece mlađe od pet godina i ukupnog rashoda u zdravstvenom sektoru u različitim zemljama?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

countries = ['United States of America', 'China', 'India', 'Nigeria', 'Germany']
df = data[data['Country'].isin(countries)][['Country', 'Total expenditure', 'under-five deaths ']]

df['Per capita expenditure'] = df['Total expenditure'] * 1000000 / data['Population']
df['Under-five mortality rate'] = df['under-five deaths '] * 1000 / data['Population']
df = df.drop(columns=['Total expenditure', 'under-five deaths '])

sns.set_style('whitegrid')
g = sns.scatterplot(x='Per capita expenditure', y='Under-five mortality rate', hue='Country', data=df, s=100, alpha=0.8, palette='bright')
plt.title('Veza između stope smrtnosti djece mlađe od pet godina i prosječnog troška po stanovniku')
plt.xlabel('Prosječan trošak po stanovniku')
plt.ylabel('Stopa smrtnosti djece mlađe od pet godina (na 1000 živorođenih)')
plt.show()
```

**Zaključak:** Sa grafika možemo zaključiti da na osnovu informacije o "under-five deaths" podataka ne možemo dobiti korisne informacije o trošku po prosjeku stanovnika. Drugim riječima, nema jasne korelacije između ovih konkretnih podataka.



**Pitanje 12:** Da li dječija paraliza utiče na očekivanu dužinu životnog vijeka?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# Odabir samo potrebnih podataka za analizu
df = df[['Country', 'Polio', 'Life expectancy ']]

# Uklanjanje nedostajućih vrijednosti
df = df.dropna()

# Grupiranje po Polio i izračunavanje srednje vrijednosti očekivane dužine života
mean_life_expectancy = df.groupby('Polio')['Life expectancy '].mean()

# Brojanje zemalja po vrijednosti kolone Polio
count_countries = df.groupby('Polio')['Country'].nunique()

# Vizualizacija podataka
plt.scatter(mean_life_expectancy.index, mean_life_expectancy.values, s=count_countries.values*10, c=count_countries.values, alpha=0.5, cmap='YlGnBu')
plt.xlabel('dječja paraliza')
plt.ylabel('Srednja vrijednost očekivane dužine života')
plt.title('Utjecaj dječje paralize na očekivanu dužinu života')
plt.colorbar(label='Broj zemalja')
plt.show()
```

**Zaključak:** Djeca koja su pogođena paralizom obično imaju smanjenu očekivanu dužinu života u poređenju sa zdravom djecom. Međutim, kada je riječ o prosječnoj dužini životnog vijeka cijele zemlje, to je kompleksno pitanje koje zavisi od mnogih faktora, a ne samo od visokog procenta djecije paralize.

Dakle, visok procenat djecije paralize može biti samo jedan od faktora koji utiču na prosječnu dužinu životnog vijeka zemlje, ali postoje i drugi faktori koji su vjerovatno važniji u tom kontekstu.

Kao što možemo primjetiti sa grafika, za veći broj zemalja, dječja paraliza ne utiče, tj. ne skraćuje dužinu životnog vijeka u prosjeku.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# scatter plot sa marginalnim distribucijama
sns.jointplot(x="Income composition of resources", y=" BMI ", data=df)
```

```python
# boxplot koristeci paket seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

#boxplot
sns.boxplot(x='Income composition of resources', y=' BMI ', data=df)
```

Kroz ova dva grafa primjetan je porast BMI-a sa porastom prihodovnog sastava resursa, porast nije linearan ali je najbliži slikom linearnoj funkciji. Ima izuzetaka koji teže nuli, tj. kada BMI teži nuli onda imamo primjere za više različitih prihodovnih sastava resursa(od 0.3 do 0.9), i obrnuto, kada prihodovni sastav resursa teži nuli tada imamo više različitih vrijednosti BMI-a(od 0 do iznad 75). Ostatak(većina) primjera su između 0.2 i 1.0 po prihodovnom sastavu resursa i između 10 i 80 po BMI-u, proporcionalno po primjerima.

Da provjerimo pretpostavku uzet ćemo i neke poznate slučajeve da usporedimo, Japan i Švicarsku kao vrlo zdrave države(dobar BMI), državu sa malim BMI-om, Madagaskar, i državu sa velikim BMI-om, Sjedinjenim Američkim Državama.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# filtriramo da koristimo samo Japan, Švicarsku, Ameriku i Madagaskar
df_selected = df[((df['Country'] == 'United States of America') | (df['Country'] == 'Japan') | (df['Country'] == 'Switzerland') | (df['Country'] == 'Madagascar')) & (df['Year'].between(2000, 2015))]

# Plot
sns.jointplot(x="Income composition of resources", y=" BMI ", data=df_selected, hue='Country')

```

Madagaskar, Japan, Švicarska pa Sjedinjene Američke Države su redom države po vrijednosti BMI-a, a tim redoslijedom su i po prihodovnom sastavu resursa.

**Zaključak:** Hipoteza se ispostavila tačna, sa vrlo malim outlierima, ima korelacija između prihodovnog sastava resursa zemlje i njenog BMI-a.

**Pitanje 15:** Da li visok procenat zastupljenosti alkohola vodi ka višoj zastupljenosti HIV/AIDS-a?

**Hipoteza:** Da, alkohol utiče na veću zastupljenost HIV/AIDS-a. Korelacija bi trebala biti vidljiva.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# scatter plot sa marginalnim distribucijama
sns.jointplot(x="Alcohol", y=" HIV/AIDS", data=df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Učitavanje podataka
df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

fig = px.scatter_3d(df, x='Alcohol', y=' HIV/AIDS', z='Life expectancy ')
fig.show()
```

Čak i kad uvedemo prosječni životni vijek ne vidimo korelaciju između zastupljenosti HIV/AIDS-a i zastupljenosti alkohola u državama, dok vidimo korelaciju između prosjećnog životnog vijeka sa i alkoholom i HIV-om, dosta "čisto".

**Zaključak:** Hipoteza nam je bila pogrešna, korelacija između zastupljenosti alkohola i HIV-a je neprimjetna, stoga je vjerovatno nema.
